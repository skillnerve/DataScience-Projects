```python
def feat(f):
  f =f
  model = RandomForestClassifier(n_estimators=400,n_jobs=-1, random_state=0)
  model.fit(Train_Data.drop([f,'Customer_ID'],axis=1), Train_Data[f])
  importance = model.feature_importances_
  features=list(Train_Data.drop('Customer_ID',axis=1).columns)
  k=0
  q=0
  w1=0
  w2=0
  for i,v in enumerate(importance):
    if str(features[i])==str(f):
        k=1
        print('*'*5)
    if v>0.01:
      if 'B2' in features[i+k].split('_'):
        if k==0:
          w1+=v
          print(features[i+k])
        else:
          w2+=v
          print(features[i+k])
      else:
        q+=v
  print(round(q,2),round(w1,2),round(w2,2))
  return(round(q,2),round(w1,2),round(w2,2))
```

```python
df=[]
for i in range(21):
  d=[]
  print('Product_Holding_B2_'+str(i))
  a,b,c=feat('Product_Holding_B2_'+str(i))
  d.append(a)
  d.append(b)
  d.append(c)
  d.append('Product_Holding_B2_'+str(i))
  df.append(d)
```
